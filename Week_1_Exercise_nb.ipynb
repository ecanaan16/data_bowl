```python
#task 1
#including important libs
import pandas as pd
import matplotlib.pyplot as plt

#converting csv files into dfs
games = pd.read_csv("/games.csv")
players = pd.read_csv("/players.csv")
plays = pd.read_csv("/plays.csv")

week_two = pd.read_csv("/week2.csv")
week_three = pd.read_csv("/week3.csv")

#reading first five rows of dfs
print(games.head())
print(players.head())
print(plays.head())
print(week_two.head())
print(week_three.head())

total_plays = week_two['playId'].nunique() + week_three['playId'].nunique()
print(f"Total number of plays: {total_plays}")

#task 2
#can create new colunn like so
#apply just performs a custom function on each row
#y does not get mirrored so it stays the same
week_three["x_std"] = week_three.apply(lambda row: 120 - row["x"] if row["playDirection"] == "left" else row["x"], axis=1)
week_three["y_std"] = week_three["y"]

#split data using groupby
#add distance between each group
#0.2, 0.4, 0.1, 0.3 -> 0.2, 0.6, 0.7, 1.0
week_three["distance"] = week_three.groupby(["gameId", "playId", "nflId"])["dis"].cumsum()

week_three["speed_mph"] = week_three["s"] * 2.045

print(week_three.head())

#task 3
#dropna removes rows that have NaN
#subset tells dropna to look at only these cols
#original df is not changed unless i include true

#use Int64 as a nullable int type in case data is missing
#becomes <NA> using pandas Int64
#col becomes a float col when NaN is involved bc NaN is a float
#so all jersey nums became floats that why we must change to Int64
#shape gives a tuple for the num of rows and cols
#actually the dropna removes the NaN and only puts everything else
#copy() is important bc sometimes dropna returns a view to the actual df and
#result in making direct changes to the original df instead of making a copy
shape_before = week_two.shape
week2_clean = week_two.dropna(subset=["x", "y", "jerseyNumber"]).copy()
week2_clean["jerseyNumber"] = week2_clean["jerseyNumber"].astype("Int64")
shape_after  = week2_clean.shape

rows_dropped = shape_before[0] - shape_after[0]

print(f"Shape before cleaning: {shape_before}")
print(f"Shape after cleaning:  {shape_after}")
print(f"Rows dropped: {rows_dropped}")

#task 4
df = week2_clean

gameId = 2018091300
playId = 181
frameId = 2

filtered_df = df[(df['gameId'] == gameId) &
                 (df['playId'] == playId) &
                 (df['frameId'] == frameId)]

```
